### why shape,size and dtypes doesnt have a bracket and others have bracket


In **Pandas**, functions or methods are followed by parentheses `()` because they are **callable**, meaning they perform an action when called. On the other hand, attributes (properties) do not have parentheses because they simply represent a value or information about the object, without performing any action.

 Heres the distinction:

#### 1. **Attributes (No Parentheses)**:
- **Attributes** provide information or describe a characteristic of the object. They do not perform any action.
- Examples: `shape`, `size`, `dtypes`
  - **`shape`**: Represents the structure of the DataFrame/Series (number of rows and columns).
  - **`size`**: Gives the total number of elements (rows  columns).
  - **`dtypes`**: Returns the data types of each column in a DataFrame or of the Series.

Since they are **attributes**, not functions, they do not need parentheses.

```python
df.shape    # (100, 5) -> Number of rows and columns
df.size     # 500 -> Total elements (100 rows * 5 columns)
df.dtypes   # Shows the data type of each column
```

#### 2. **Methods/Functions (With Parentheses)**:
- **Methods** are **callable functions** that perform an operation or computation on the DataFrame or Series.
- Examples: `head()`, `mean()`, `sum()`, `describe()`
  - **`head()`**: Retrieves the first n rows.
  - **`mean()`**: Computes the mean (average) of the values.
  - **`describe()`**: Generates descriptive statistics.

Since they are **callable**, they require parentheses to execute.

```python
df.head()        # Retrieves the first 5 rows
df.mean()        # Calculates the mean of each column
df.describe()    # Provides a summary of statistics
```

**Summary:**
- **Attributes**: Provide information (e.g., `shape`, `size`, `dtypes`). No parentheses needed.
- **Methods**: Perform operations or actions (e.g., `head()`, `mean()`, `describe()`). Parentheses required.

Attributes give you direct information, while methods execute actions on the data.

You can download the data from the provided API URL using Python with the help of the **`requests`** library to fetch the data and the **`pandas`** library to process the CSV data.

Here's a step-by-step guide to download the data using Python:

### Steps:
1. **Use the `requests` library** to send an HTTP request to the API URL and get the CSV data.
2. **Use `pandas`** to load and save the CSV data to a local file.

- **Explanation**
1. **`requests.get()`**: This sends a GET request to the provided API URL.
2. **`response.text`**: Contains the raw CSV data from the API.
3. **`pd.read_csv()`**: Reads the CSV content into a Pandas DataFrame.
4. **`data.to_csv()`**: Saves the DataFrame to a CSV file locally.
5. **`status_code == 200`**: Checks if the API request was successful.

````python
import requests
import pandas as pd
import io

# API URL for downloading the CSV data
api_url = 'https://api.data.gov.in/resource/9fed9e21-d687-4c24-aa6e-63cb24bb68d3?api-key=579b464db66ec23bdd000001cdd3946e44ce4aad7209ff7b23ac571b&format=csv'

# Send a GET request to the API
response = requests.get(api_url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Convert the response content to a pandas DataFrame using io.StringIO
    data = pd.read_csv(io.StringIO(response.text))
    
    # Save the data to a local CSV file
    data.to_csv('downloaded_data.csv', index=False)
    
    print("Data downloaded and saved as 'downloaded_data.csv'")
else:
    print(f"Failed to download data. HTTP Status code: {response.status_code}")
````
